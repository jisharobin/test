```python
# Find function with path compression
def find(x, parent):
    if parent[x] == 0:
        return x
    parent[x] = find(parent[x], parent)  # Path compression
    return parent[x]

# Union function
def union(x, y, parent):
    root_x = find(x, parent)
    root_y = find(y, parent)
    if root_x != root_y:
        parent[root_y] = root_x  # Make root_x the parent of root_y

# Example usage
parent = [0, 2, 3, 0, 3, 7, 7, 0, 3]

# Find the root of 1
print(find(1, parent))  # Output: 3

# Union sets containing 1 and 5
union(1, 5, parent)
print(parent)  # Updated parent array
```

```python
import cv2
import numpy as np

# Load or create binary images
# Seed image (initial set)
seed = np.zeros((10, 10), dtype=np.uint8)
seed[4, 4] = 1  # A single starting point for dilation


# Mask image (constraint set)
mask = np.zeros((10, 10), dtype=np.uint8)
mask[2:8, 2:8] = 1  # A square mask with boundaries
mask[6, 6] = 0       # Adding a hole in the mask to constrain dilation


# Structuring element
structuring_element = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))

# Conditional dilation function
def conditional_dilation(seed, mask, struct_element, max_iter=100):
    result = seed.copy()
    for i in range(max_iter):
        # Perform dilation on the current result
        dilated = cv2.dilate(result, struct_element)

        # Apply the mask (constraint)
        new_result = cv2.bitwise_and(dilated, mask)

        # Check for stability (if nothing changes, stop)
        if np.array_equal(result, new_result):
            break

        result = new_result

    return result

# Perform conditional dilation
result = conditional_dilation(seed, mask, structuring_element)

# Print and visualize the results
print("Seed Image:\n", seed)
print("Mask Image:\n", mask)
print("Result After Conditional Dilation:\n", result)
```
